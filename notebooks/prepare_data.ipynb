### Prepare COVID data

First rename columns, then drop those that have any NAs

```python
import pandas as pd
from pathlib import Path

data_path = Path("../data")

df_covid = pd.read_csv(data_path / "covid.csv")

print(f"Original size: {df_covid.shape[0]}")

df_covid = df_covid.rename(columns={
    'Unnamed: 0': 'id',
    'text_x': 'context',
    'text_y': 'question'
})

df_covid = df_covid.dropna()

df_covid
```

```python
df_covid.to_csv(data_path / "covid_new.csv", index=False)
```
